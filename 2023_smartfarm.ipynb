```python
input_data = input_data.drop(columns=['frmDist'])

input_data
```

```python
input_data["date"] = pd.to_datetime(input_data["date"],format='%Y%m%d')
# input_data['year']= input_data['date'].dt.year
input_data['month']= input_data['date'].dt.month
# input_data['day']= input_data['date'].dt.day
input_data
```

```python
#계절 변수를 미리 생성
input_data['season']='season'

input_data.loc[(input_data['month']>=2) & (input_data['month']<5) ,['season']]= 'spring'
input_data.loc[(input_data['month']>=5) & (input_data['month']<8) ,['season']]= 'summer'
input_data.loc[(input_data['month']>=8) & (input_data['month']<11),['season']]= 'autumn'
input_data.loc[(input_data['month']>=11) & (input_data['month']<13) & (input_data['month']==1),['season']]= 'winter'
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from category_encoders import OneHotEncoder
encoder = OneHotEncoder(use_cat_names = True)
df_onehot = encoder.fit_transform(input_data[['season']])
```

```python
df_onehot
```

```python
input_ohe = pd.concat([input_data, df_onehot],axis=1)
# input_ohe = input_data.drop(columns=['season'])
```

```python
input_ohe = input_ohe.drop(columns=['season'])
input_ohe
```

```python
# Split the data into training and testing sets
X = input_ohe[input_ohe.drop(columns=['date','outtrn_cumsum','HeatingEnergyUsage_cumsum']).columns]
Y = input_ohe[['outtrn_cumsum','HeatingEnergyUsage_cumsum']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,random_state=121)
```

```python
# Calculate RMSE between the predictions and actual 'y' values
def calculate_rmse(targets, predictions):
    """
    Calculate the Root Mean Squared Error (RMSE) between predicted and target values.

    :param predictions: Predicted values.
    :type predictions: array-like
    :param targets: Target values.
    :type targets: array-like
    :return: RMSE value.
    :rtype: float
    """
    from sklearn.metrics import mean_squared_error
    return np.sqrt(mean_squared_error(targets, predictions))


# Calculate r2_score between the predictions and actual 'y' values
def calculate_R2_score(y_test,y_pred):
    from sklearn.metrics import r2_score
    return r2_score(y_test, y_pred)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import ExtraTreesRegressor  # Import ExtraTreesRegressor
from sklearn.multioutput import MultiOutputRegressor


# Create an ExtraTreesRegressor with MultiOutputRegressor for multi-output regression
model = MultiOutputRegressor(ExtraTreesRegressor())  # Change to ExtraTreesRegressor

# Fit the model on the training data
model.fit(X_train, y_train)

# Predict on the test data
y_pred = model.predict(X_test)

# Calculate RMSE and R2_score
rmse = calculate_rmse(y_test, y_pred)
r2score = calculate_R2_score(y_test, y_pred)


# ------------------------------------------------
### OUTPUT ###
print("RMSE:", rmse)
print("R2_score:", r2score)
```
